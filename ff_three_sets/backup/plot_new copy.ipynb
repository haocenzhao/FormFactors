### John's GE/GD paramaterizated uncertainty

```python
##John's GE/GD paramaterizated uncertainty
def GEGD_Err(Qsq):
    p1=1.0
    p2=2.36527938
    p3=-3.34619e-2
    p4=4.2117
    p5=1.4510
    p6=5.7528
    p7=0.65085
    p8=3.2961

    x=np.log(Qsq)
    dge=sqrt(min(exp(p1*x-p2),exp(p3*x-p4))**2+min(exp(p5*x-p6),exp(p7*x-p8))**2)
    return dge
```

### John's GM/mu/GD paramaterizated uncertainty

```python
##John's GM/mu/GD paramaterizated uncertainty
def GMGD_Err(Qsq):
    p1=1.0
    p2=1.6749              
    p3=0.53139              
    p4=3.9545              
    p5=0.59457              
    p6=1.5804              
    p7=-4.5913              
    p8=-4.0121              
    p9=-2.4233              
    p10=2.1316              
    p11=2.8336              
    p12=1.1837              

    x=np.log(Qsq)
    dgm=sqrt(min(exp(p1*x-p2),exp(p3*x-p4),exp(min(p5*abs(x-p6)**p12+p7,p8*x/x)))**2+(exp(p9+p10*log(max(1e-4,x-p11))))**2)

    return dgm
```

## Bosted 94 - Proton and Neutron

$$G_E^p(Q^2) = \frac{1}{1+0.62 Q + 0.68 Q^2 + 2.80 Q^3 + 0.83 Q^4} $$,
$$G_M^p/\mu_p = \frac{1}{1+0.35 Q + 2.44 Q^2 + 0.50 Q^3 + 1.04 Q^4 + 0.34 Q^5}$$
$$G_M^n/\mu_n = \frac{1}{1-1.74 Q + 9.29 Q^2 - 7.63 Q^3 + 4.64 Q^4} $$

```python
N=2000
GEp_PB94 = np.zeros(N, dtype=float)
GMp_PB94 = np.zeros(N, dtype=float)
GEpGD_PB94 = np.zeros(N, dtype=float)
GMpGD_PB94 = np.zeros(N, dtype=float)
GEn_PB94 = np.zeros(N, dtype=float)
GMn_PB94 = np.zeros(N, dtype=float)
GEnGD_PB94 = np.zeros(N, dtype=float)
GMnGD_PB94 = np.zeros(N, dtype=float)
for i in range(0,N):
    GEDp_tmp = GD(Q2[i]) * GEp0
    GMDp_tmp = GD(Q2[i]) * GMp0
    GEDn_tmp = GD(Q2[i]) * GEn0
    GMDn_tmp = GD(Q2[i]) * GMn0  
    Q=sqrt(Q2[i])
    
    GEp_PB94[i] = 1./(1.+0.62*Q+0.68*Q**2+2.80*Q**3+0.83*Q**4)
    GMp_PB94[i] = 1./(1.+0.35*Q+2.44*Q**2+0.50*Q**3+1.04*Q**4+0.34*Q**5)*GMp0
    GMn_PB94[i] = 1./(1.-1.74*Q+9.29*Q**2-7.63*Q**3+4.63*Q**4)*GMn0
    tau = Q2[i]/4.0/Mass_n**2
    GEn_PB94[i] = -1.25*GMn0*tau/(1.+18.3*tau)*GD(Q2[i])
    
    GEpGD_PB94[i]=GEp_PB94[i]/GEDp_tmp
    GMpGD_PB94[i]=GMp_PB94[i]/GMDp_tmp
    GEnGD_PB94[i]=GEn_PB94[i]/GEDn_tmp
    GMnGD_PB94[i]=GMn_PB94[i]/GMDn_tmp
```

## Friendrich and Walcher 03 - Proton and Neutron

$$G_N(Q^2)=G_s(Q^2)+a_b Q^2 G_b(Q^2)$$, where,
$G_s(Q^2)=\frac{a_{10}}{(1+Q^2/a_{11})^2}+\frac{a_{20}}{(1+Q^2/a_{21})^2}$, and,
$G_b(Q^2)=e^{-\frac{1}{2}(\frac{Q-Q_b}{\sigma_b})^2}+e^{-\frac{1}{2}(\frac{Q+Q_b}{\sigma_b})^2}$

Here are the fitting parameters:
<img src="F_W_Table.png">


```python
def GN(kQsq, a10, a11, a20, a21, ab, Qb, sigb):
    Gs = a10/(1.+kQsq/a11)**2 + a20/(1.+kQsq/a21)**2
    Gb = exp(-0.5*((sqrt(kQsq)-Qb)/sigb)**2)+exp(-0.5*((sqrt(kQsq)+Qb)/sigb)**2)
    Gn = Gs + ab * kQsq * Gb
    return Gn

## Start calcualting
N=2000
GEp_FW03 = np.zeros(N, dtype=float)
GMp_FW03 = np.zeros(N, dtype=float)
GEpGD_FW03 = np.zeros(N, dtype=float)
GMpGD_FW03 = np.zeros(N, dtype=float)
GEn_FW03 = np.zeros(N, dtype=float)
GMn_FW03 = np.zeros(N, dtype=float)
GEnGD_FW03 = np.zeros(N, dtype=float)
GMnGD_FW03 = np.zeros(N, dtype=float)
for i in range(0,N):
    GEDp_tmp = GD(Q2[i]) * GEp0
    GMDp_tmp = GD(Q2[i]) * GMp0
    GEDn_tmp = GD(Q2[i]) * GEn0
    GMDn_tmp = GD(Q2[i]) * GMn0  

    GEp_FW03[i] = GN(Q2[i], 1.041, 0.765, -0.041, 6.2, -0.23, 0.07, 0.27)
    GMp_FW03[i] = GN(Q2[i], 1.002, 0.749, -0.002, 6.0, -0.13, 0.35, 0.21) * GMp0
    GEn_FW03[i] = GN(Q2[i], 1.040, 1.730, -1.040, 1.54, 0.23, 0.29, 0.20)
    GMn_FW03[i] = GN(Q2[i], 1.012, 0.770, -0.012, 6.8, -0.28, 0.33, 0.14) * GMn0 
    
    GEpGD_FW03[i]=GEp_FW03[i]/GEDp_tmp
    GMpGD_FW03[i]=GMp_FW03[i]/GMDp_tmp
    GEnGD_FW03[i]=GEn_FW03[i]/GEDn_tmp
    GMnGD_FW03[i]=GMn_FW03[i]/GMDn_tmp    
```

## BBBA 06 - Proton and Neutron
  $G_E, G_M/\mu_p = \frac{\sum_{k=0}^n a_k \tau^k}{1+\sum_{k=1}^{n+1} b_k \tau^k}$, where $\tau = Q^2/4M$.
Here are the fitting parameters:
<img src="BBBA_Table.png">

```python
def GNFW(kTau, a0, a1, a2, b1, b2, b3, b4):
    Gn = (a0 + a1*kTau + a2*kTau**2)/(1.+b1*kTau**1+b2*kTau**2+b3*kTau**3+b4*kTau**4)
    return Gn

##Start calculating
N=2000
GEp_BB06 = np.zeros(N, dtype=float)
GMp_BB06 = np.zeros(N, dtype=float)
GEpGD_BB06 = np.zeros(N, dtype=float)
GMpGD_BB06 = np.zeros(N, dtype=float)
GEn_BB06 = np.zeros(N, dtype=float)
GMn_BB06 = np.zeros(N, dtype=float)
GEnGD_BB06 = np.zeros(N, dtype=float)
GMnGD_BB06 = np.zeros(N, dtype=float)
for i in range(0,N):
    GEDp_tmp = GD(Q2[i]) * GEp0
    GMDp_tmp = GD(Q2[i]) * GMp0
    GEDn_tmp = GD(Q2[i]) * GEn0
    GMDn_tmp = GD(Q2[i]) * GMn0  

    tau = Q2[i]/4.0/Mass_p**2
    GEp_BB06[i] = GNFW(tau, 1., -0.0578, 0.0, 11.1, 13.6, 33.0, 0.0)
    GMp_BB06[i] = GNFW(tau, 1.,  0.150,  0.0, 11.1, 19.6, 7.54, 0.0) * GMp0

    tau = Q2[i]/4.0/Mass_n**2
    GEn_BB06[i] = GNFW(tau, 0.0, 1.25, 1.30, -9.86, 305.0, -758.0, 802.0)
    GMn_BB06[i] = GNFW(tau, 1.0, 1.81, 0.00, 14.10,  20.7,   68.7, 0.0) * GMn0 
    
    GEpGD_BB06[i]=GEp_BB06[i]/GEDp_tmp
    GMpGD_BB06[i]=GMp_BB06[i]/GMDp_tmp
    GEnGD_BB06[i]=GEn_BB06[i]/GEDn_tmp
    GMnGD_BB06[i]=GMn_BB06[i]/GMDn_tmp    
```

# Flavor Separation
If only considering the u and d quark contributions, we have:
$$G^p_E = 2/3 G^u_E  - 1/3 G^d_E, G^n_E = 2/3 G^d_E - 1/3 G^u_E,$$
which gives:
$$G^u_E = 2G^p_E + G^n_E,\delta G^u_E = \sqrt{(2\delta G^p_E)^2+(\delta G^n_E)^2 }$$
$$G^d_E = 2G^n_E + G^p_E,\delta G^d_E = \sqrt{(2\delta G^n_E)^2+(\delta G^p_E)^2 }.$$

Similarly, for magnetic form factors, we just replace "E" by "M". 
   
   $F_{1/2}^{p/n}$ are calculated based on: 
    $$F_1^p =  \frac{G_E^p + \frac{Q^2}{4 M_p^2} G_M^p}{1+\frac{Q^2}{4 M_p^2}},  
    \delta F_1^p =  \frac{\sqrt{(\delta G_E^p)^2 + (\frac{Q^2}{4 M_p^2} \delta G_M^p)^2}}{1+\frac{Q^2}{4 M_p^2}}$$
    $$ F_2^p = -\frac{G_E^p - G_M^p}{ 1+\frac{Q^2}{4 M_p^2}}, 
    \delta F_2^p =  \frac{\sqrt{(\delta G_E^p)^2 + (\delta G_M^p)^2}}{1+\frac{Q^2}{4 M_p^2}}$$

    
A similar approach was performed to extract $F_{1/2}^{u/d}$ like $G_{E/M}^{p/n}$
