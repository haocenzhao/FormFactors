```python
##Import and Definition
import numpy as np
from math import *
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.ticker import FixedLocator, MultipleLocator, FormatStrFormatter
import os, sys, traceback
import matplotlib.gridspec as gridspec

#very import to enable this so the plots can be showed in the page
get_ipython().magic(u'matplotlib inline')

# Use LaTeX font.
plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['Computer Modern Roman'],'size':20})
```

### Load Old Results

```python
##Load Old Results{{{
##################################################
import FF_load
fitlines = open('World_CrossSections.dat','r').readlines()
N = 7
Q2_h = np.zeros(N,dtype=float)
x_h = np.zeros(N,dtype=float)
y_h = np.zeros(N,dtype=float)
z_h = np.zeros(N,dtype=float)
pt_h = np.zeros(N,dtype=float)
Asym_h = np.zeros(N,dtype=float)
Astat_h = np.zeros(N,dtype=float)
Asyst_h = np.zeros(N,dtype=float)
Aerr_h = np.zeros(N,dtype=float)

sum_syst = 0.0
sum_stat = 0.0
sum_tot=0.0
for i in range(0,7):
    values = fitlines[i+15].split()
    Q2_h[i] = float(values[0])
    x_h[i] = float(values[1])
    y_h[i] = float(values[2])
    z_h[i] = float(values[3])
    pt_h[i] = float(values[4])
    Asym_h[i] = float(values[5])
    Astat_h[i] = float(values[6])
    Asyst_h[i] = float(values[7])    
    Aerr_h = sqrt(Astat_h[i]**2 + Asyst_h[i]**2)
    
    sum_stat += Astat_h[i]/Asym_h[i] *100.
    sum_syst += Asyst_h[i]/Asym_h[i] *100.
    sum_tot += Aerr_h/Asym_h[i] * 100.
    #sum_stat += Astat_h[i] *100.
    #sum_syst += Asyst_h[i] *100.
    #sum_tot += Aerr_h * 100.
    
print "average HERMES uncertainties: syst=%f, stat=%f, tot=%f"%(sum_syst/N, sum_stat/N, sum_tot/N)
    #print 'Q2=%f, x=%f'%(Q2_h[i], x_h[i])
```

### Load SoLID $^{3}He$(e,e'$\pi^{+}$)X projection data

```python
fitlines = open('3he_pip_hemes.dat','r').readlines()
N = len(fitlines)-1
Q2_s = np.zeros(N,dtype=float)
x_s = np.zeros(N,dtype=float)
y_s = np.zeros(N,dtype=float)
z_s = np.zeros(N,dtype=float)
pt_s = np.zeros(N,dtype=float)
Asym_s = np.zeros(N,dtype=float)
Astat_s = np.zeros(N,dtype=float)
Asyst_s = np.zeros(N,dtype=float)
Aerr_s = np.zeros(N,dtype=float)

sum_syst = 0.0
sum_stat = 0.0
sum_tot=0.0
for i in range(0,N-1):
    values = fitlines[i+1].split()
    Q2_s[i] = float(values[0])
    x_s[i] = float(values[1])
    y_s[i] = float(values[2])
    z_s[i] = float(values[3])
    pt_s[i] = float(values[4])
    Asym_s[i] = float(values[5])
    Astat_s[i] = float(values[6])
    Asyst_s[i] = float(values[7])   
    Aerr_s = sqrt(Astat_s[i]**2 + Asyst_s[i]**2)
    sum_stat += Astat_s[i] *100.
    sum_syst += Asyst_s[i] *100.
    sum_tot += Aerr_s * 100.
    #print 'Q2=%f, x=%f'%(Q2_s[i], x_s[i])
print "average SoLID uncertainties: syst=%f, stat=%f, tot=%f"%(sum_syst/N, sum_stat/N, sum_tot/N)
```
