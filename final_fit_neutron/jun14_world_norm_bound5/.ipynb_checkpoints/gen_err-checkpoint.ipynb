# Richard's Parameterization

$$ log_{10}\frac{\delta G}{G_D} = (L+c_0)\Theta_a(L_1-L) 
                                +\sum_{i=1}^{N}(c_i+d_i L)[\Theta_a(L_i-L)-\Theta_a(L_{i+1}-L)]
                                +log_{10}(E_{\inf})\Theta_a(L-L_{N+1})$$
where $L=log_{10}(Q^2)$, $\Theta_{a}(x)=[1+10^{-ax}]^{-1}$. $a=1$.

```python
## Define Functions
def THETA(x):
    th = 1./(1.+10**(-x))
    return th

## Richard's parameterized formula
def dG_RH(kQ2, Einf,c0,c, d, Lp):
    n = len(c)
    np = len(d)
    npp = len(Lp)
    if n!=np or n!=npp-1:
        print '*** ERROR, the length of "c"(%d) or "d"(%d) or "L"(%d) are not equal!!!'%(n, np, npp)
        
    L = log10(kQ2)  
    term1 = (L+c0)*THETA(Lp[0]-L)
    term3 = log10(Einf)*THETA(L-Lp[npp-1])
    term2 = 0.0
    for i in range(0,n):
        term2 += (c[i]+d[i]*L) * ( THETA(Lp[i]-L) - THETA(Lp[i+1]-L) )
    
    ##LOG10(dG/GD)
    dG_GD = (term1+term2+term3) 
    ##Note: return dG/GD here
    return (10.**dG_GD) 
```

# Calculating Errors 

```python
## Model Dependent Errors by using different data sets

GE_err_fit = np.zeros(len(Q2e_n), dtype=float)
GM_err_fit = np.zeros(len(Q2e_n), dtype=float)
GE_err_sum = np.zeros(len(Q2e_n), dtype=float)
GM_err_sum = np.zeros(len(Q2e_n), dtype=float)
GE_err_rh = np.zeros(len(Q2e_n), dtype=float)
GM_err_rh = np.zeros(len(Q2e_n), dtype=float)

of = open('neutron_baseline.dat','w')
print >> of, '##%10s %12s %14s %14s %14s %14s %14s %14s'%('Q2','Z','GEn/GD','dGEn/GD','dGEn_Par/GD','GMn/muGD','dGMn/muGD','dGMn_Par/muGD')
for i in range(0, len(Q2e_n)):    
        GE_err_fit[i] = abs((GE_pos_n[i]-GE_fit_n[i])/GED_n[i])
        GM_err_fit[i] = abs((GM_pos_n[i]-GM_fit_n[i])/GMD_n[i])
        GE_err_sum[i] = abs((GE_pos_n1[i]-GE_fit_n1[i])/GED_n1[i])
        GM_err_sum[i] = abs((GM_pos_n1[i]-GM_fit_n1[i])/GMD_n1[i])

        #GE_err_rh[i] = dG_RH(Q2e_n[i], Einf_GEn, c0_GEn5, c_GEn5, d_GEn5, L_GEn5 ) /abs(GE0)
        #GM_err_rh[i] = dG_RH(Q2e_n[i], Einf_GMn, c0_GMn5, c_GMn5, d_GMn5, L_GMn5 ) /abs(GM0)
        
        GE_err_rh[i] = dG_RH(Q2e_n[i], Einf_GEn, c0_GEn, c_GEn, d_GEn, L_GEn ) /abs(GE0)
        GM_err_rh[i] = dG_RH(Q2e_n[i], Einf_GMn, c0_GMn, c_GMn, d_GMn, L_GMn ) /abs(GM0)
        
        print >> of, '%12s %12s %14.6e %14.6e %14.6e %14.6e %14.6e %14.6e'%(Q2e_n[i],Ze_n[i],GE_fit_n[i]/GED_n[i], GE_err_sum[i], GE_err_rh[i],GM_fit_n[i]/GMD_n[i], GM_err_sum[i], GM_err_rh[i]  )
of.close()
  
        
```
